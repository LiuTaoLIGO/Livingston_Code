```python
kat2 = LLO.deepcopy()
kat2.DARM.DC = 0.0033834586466165405/2 # So that the DARM offset is 10 pm.
R0 = 36.027
delta1 = 0.001
delta2 = 0.0005

kat2.parse(f"""
pd Pin PRMAR.p1.i
pd Pcirc PRM.p2.o

mmd modeMismatchx ITMX.p1.i ITMX.p2.o direction=y
mmd modeMismatchy ITMY.p1.i ITMY.p2.o direction=x

xaxis(PR3.Rcx, "lin", {R0*(1-delta1)}, {R0*(1+delta2)}, 800, pre_step=change(PR3.Rcy=PR3.Rcx))
""")

outLLO2 = kat2.run()
PRG = outLLO2['Pcirc']/outLLO2['Pin']

xaxis = np.linspace(-delta1, delta2, 801)*100

fig, ax = plt.subplots(1, 1, figsize=(7,5))

lns1 = ax.plot(xaxis, PRG, color="g", label="PRG")

ax.set_ylabel("PRG", fontsize=15)
ax.set_xlabel(r"$\Delta$RoC for PR3 [%]", fontsize=13)

ax_2 = ax.twinx()
lns2 = ax_2.semilogy(xaxis, outLLO2["modeMismatchx"], color="b", label="Mode Mismatch at ITMX")
lns3 = ax_2.semilogy(xaxis, outLLO2["modeMismatchy"], "r--", label="Mode Mismatch at ITMY")

ax_2.set_ylabel("Mode Mismatch", fontsize=15)

ax_2.grid(visible=False, which="minor", axis='y')

# ax_2.legend(loc = 'best', fontsize=13)

lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc='best', fontsize=13)

value = (36-R0)/R0*100
idx = np.abs(xaxis-value).argmin()
print(f"When the RoC for PR3 is 36 m, the PRG is {PRG[idx]:.6f}")
print()
print(f"PRG maximum happens at when RoC equals to {xaxis[PRG.argmax()]:.4f} %")
print(f"Mode Mismatch minimum happens at RoC equals to {xaxis[outLLO2['modeMismatchx'].argmin()]:.4f} %")
fig.savefig('PRG.pdf', bbox_inches='tight', dpi=120, pad_inches=0.01)
```
