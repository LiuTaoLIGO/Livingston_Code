## Scipy minimize to find the losses in the arms as well as the PR2-PR3 length, in order to match the PRG, REFL and AS total power

```python
katOPT = LLO.deepcopy()
katOPT.DARM.DC = 0.0014 
katOPT.remove(katOPT.DARM_dc_lock)

katOPT.parse("""
pd Pin PRMAR.p1.i
pd PRefl PRMAR.p1.o
pd PPoP PR2.p3.o
pd PAS SRMAR.p2.o

ad ASC SRMAR.p2.o f=0

# For PRG
pd Pcirc PRM.p2.o
""")
```

```python
katOPT.ITMX.Rc
```

```python
def objective_function(params):
    lossX, lossY, pr2pr3L, ITMYRc, ITMXRc = params

    kat = katOPT.deepcopy()
    kat.ETMY.set_RTL(L=4e-05+lossY, T=3.9e-06)
    kat.ETMX.set_RTL(L=4e-05+lossX, T=4e-06)
    kat.lp2.L = 16.1648*(1+pr2pr3L)
    kat.ITMY.Rc = [-1940.7*(1+ITMYRc), -1940.7*(1+ITMYRc)]
    kat.ITMX.Rc = [-1937.9*(1+ITMXRc), -1937.9*(1+ITMXRc)]
    


    out = kat.run(Noxaxis(pre_step=RunLocks(method='newton', 
                                            display_progress=False,
                                            max_iterations=1000)))

    return np.abs(np.array([out['Pcirc']/out['Pin']-42.5, out['PAS']-0.533, out['PRefl']-1.534])).sum()


initial_guess = (1e-05, 1e-05, 1e-05, 1e-05, 1e-05)
result = optimize.minimize(
                        objective_function, 
                        initial_guess, 
                        method='Nelder-Mead',
                        options={'maxiter':1e6},
                        bounds=[(-4e-05, 1-4e-05-4e-06), (-4e-05, 1-4e-05-4e-06), (-1,1), (-10,10), (-10,10)], 
                        tol=1e-9)
if result.success:
    fitted_params = result.x
    print(fitted_params)
else:
    raise ValueError(result.message)
```

```python
a = 0.17 # mirror radius
h = 0.2  # mirror thickness
wa = 47e-6/2 # absorber size
wb = 41e-6/2 # absorber size
Pa = 19.5e-3 # power absorbed
Pb = 13.7e-3 # power absorbed

xa, ya = [14.6e-3, 43.9e-3]
xb, yb = [-47.4e-3, -7.1e-3]

xs = ys = np.linspace(-a, a, 300) # radial points

xsa = np.linspace(-a-xa, a-xa, 300)
ysa = np.linspace(-a-ya, a-ya, 300)

xsb = np.linspace(-a-xb, a-xb, 300)
ysb = np.linspace(-a-yb, a-yb, 300)

spa_a = spa(xsa, ysa, wa, h, Pa)
spa_b = spa(xsb, ysb, wb, h, Pb)

Pmain = 5e-3
wmain = 24e-6/2
xmai, ymai = [5e-3, -35e-3]

Pnew = 5e-3  # a new point absorber on the ETMY
wnew = 24e-6/2
xne, yne = [20e-3, -20e-3]

xnew = np.linspace(-a-xne, a-xne, 300)
ynew = np.linspace(-a-yne, a-yne, 300)
spa_new = spa(xnew, ynew, wnew, h, Pnew)

xmain = np.linspace(-a-xmai, a-xmai, 300)
ymain = np.linspace(-a-ymai, a-ymai, 300)
spa_main = spa(xmain, ymain, wmain, h, Pmain)

spa_ETMX = spa_a + spa_b
spa_ETMY = spa_main + spa_new # a new point absorber on the ETMY


fig = plt.figure(figsize=(10, 6))
ax1 = fig.add_subplot(121)
im1 = ax1.imshow(spa_ETMY, interpolation='None')
ax1.set_title(f"ETMY")
ax1.grid(visible=None)

divider = make_axes_locatable(ax1)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im1, cax=cax, orientation='vertical')

ax2 = fig.add_subplot(122)
im2 = ax2.imshow(spa_ETMX, interpolation='None')
ax2.set_title(f"ETMX")

ax2.grid(visible=None)

divider = make_axes_locatable(ax2)
cax = divider.append_axes('right', size='5%', pad=0.05)
fig.colorbar(im2, cax=cax, orientation='vertical');

# fig.savefig('./pointAbsorbersETM.pdf', bbox_inches='tight', dpi=180, pad_inches=0.01)
```
